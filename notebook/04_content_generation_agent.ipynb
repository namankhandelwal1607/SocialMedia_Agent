```python
from pydantic import BaseModel
from typing import List
```

```python
class SocialPost(BaseModel):

    platform: str

    title: str

    content: str

    hashtags: list[str]
```

```python
from dotenv import load_dotenv
load_dotenv()

from langchain_groq import ChatGroq

Groq = ChatGroq(
    model="llama-3.3-70b-versatile",
)
```

```python
from langchain.agents import create_agent
content_agent = create_agent(
    model=Groq,
    response_format=SocialPost,
    system_prompt="""
You are a Social Media Content Generation Agent.

Your job is to create engaging,
professional social media content.

Guidelines:

- Follow the provided strategy
- Maintain Auriga's professional tone
- Avoid exaggerated marketing claims
- Focus on thought leadership
- Generate relevant hashtags
"""
)
```

```python
response = content_agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": """
                
Trend:
Agentic AI

Auriga Knowledge:
Auriga helps enterprises adopt AI,
data analytics and digital transformation.

Strategy:

Audience:
Business Leaders

Platform:
LinkedIn

Content Type:
Article

Angle:
How Agentic AI is Revolutionizing Business Operations

CTA:
Consult with Auriga Experts to Leverage Agentic AI
                """
            }
        ]
    }
)
```

```python
post = response["structured_response"]

print(post.title)

print("\n")

print(post.content)

print("\n")

print(post.hashtags)
```
