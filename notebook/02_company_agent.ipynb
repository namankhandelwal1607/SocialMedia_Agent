```python
IMPORTANT_URLS = [
    # Company
    "https://aurigait.com/",
    "https://aurigait.com/about-us/",
    "https://aurigait.com/about-us/our-journey/",
    "https://aurigait.com/about-us/our-team/",

    # Services
    "https://aurigait.com/services/artificial-intelligence/",
    "https://aurigait.com/services/cloud-devops/",
    "https://aurigait.com/services/data-analytics/",
    "https://aurigait.com/services/digital-products-and-software/",
    "https://aurigait.com/services/erp-solutions/",
    "https://aurigait.com/services/user-experience-design/",
    "https://aurigait.com/services/enterprise-business-platforms/",

    # Products
    "https://aurigait.com/products-solutions/cygnus-alpha/",
    "https://aurigait.com/products-solutions/instadigin/",
    "https://aurigait.com/products-solutions/safira/",
    "https://aurigait.com/products-solutions/stitch-mes/",

    # AI
    "https://aurigait.com/ai-readiness-assessment/",
    "https://aurigait.com/our-work/?category=artificial-intelligence",

    # Case Studies
    "https://aurigait.com/case-study/ai-powered-traffic-management-system-for-national-highways/",
    "https://aurigait.com/case-study/automating-central-bank-compliant-loan-document-generation-for-all-loan-types/",
    "https://aurigait.com/case-study/erpnext-manufacturing-erp-multi-plant/",
    "https://aurigait.com/case-study/improving-the-sellers-marketplace-experience/",
    "https://aurigait.com/case-study/launched-multiple-digital-services-for-yes-bank/",
]
```

```python
from langchain_core.documents import Document
```

```python
def scrape_page(url):

    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    response = requests.get(
        url,
        headers=headers,
        timeout=30
    )

    soup = BeautifulSoup(
        response.text,
        "html.parser"
    )

    # Remove noisy tags
    for tag in soup([
        "script",
        "style",
        "noscript"
    ]):
        tag.decompose()

    text = soup.get_text(
        separator="\n",
        strip=True
    )

    return Document(
        page_content=text,
        metadata={
            "source": url
        }
    )
```

```python
documents = []

for url in IMPORTANT_URLS:

    try:

        doc = scrape_page(url)

        documents.append(doc)

        print(f"✓ {url}")

    except Exception as e:

        print(f"✗ {url}")
        print(e)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = text_splitter.split_documents(
    documents
)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(
    chunks,
    embeddings
)
```

```python
vectorstore.save_local(
    "../vectorstore/auriga"
)
```

```python
vectorstore = FAISS.load_local(
    "../vectorstore/auriga",
    embeddings,
    allow_dangerous_deserialization=True
)
```

```python
retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={
        "k": 5,
        "fetch_k": 20
    }
)
```

```python
results = retriever.invoke(
    "What AI services does Auriga provide?"
)
```

```python
for doc in results:

    print(doc.metadata)

    print(doc.page_content[:1000])

    print("=" * 100)
```

```python
queries = [
    "What AI services does Auriga provide?",
    "What products does Auriga offer?",
    "What is Cygnus Alpha?",
    "What ERP solutions does Auriga provide?",
    "What AI case studies has Auriga worked on?"
]

for query in queries:
    print("\n", "="*100)
    print("QUERY:", query)

    docs = retriever.invoke(query)

    print(docs[0].metadata)
    print(docs[0].page_content[:500])
```

```python
from langchain_core.tools import tool

@tool
def auriga_search(query: str):
    """
    Search Auriga knowledge base.
    """

    docs = retriever.invoke(query)

    return "\n\n".join(
        doc.page_content
        for doc in docs
    )
```

```python
auriga_search.invoke(
    "What AI services does Auriga provide?"
)
```

```python
from dotenv import load_dotenv
load_dotenv()

from langchain_groq import ChatGroq

Groq = ChatGroq(
    model="llama-3.3-70b-versatile",
)
```

```python
from langchain.agents import create_agent

knowledge_agent = create_agent(
    model=Groq,
    tools=[auriga_search],
    system_prompt="""
    You are Auriga's Knowledge Agent.

    Your job is to answer questions about:
    - Auriga services
    - Products
    - Case studies
    - AI offerings
    - Industry expertise

    Always use the knowledge base.
    """
)
```

```python
response = knowledge_agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "How is Auriga related to Agentic AI?"
            }
        ]
    }
)
```
