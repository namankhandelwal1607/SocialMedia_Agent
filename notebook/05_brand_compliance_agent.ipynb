```python
from pydantic import BaseModel
class ComplianceReview(BaseModel):

    approved: bool

    feedback: str

    risk_level: str
```

```python
from dotenv import load_dotenv
load_dotenv()

from langchain_groq import ChatGroq

Groq = ChatGroq(
    model="llama-3.3-70b-versatile",
)
```

```python
from langchain.agents import create_agent
compliance_agent = create_agent(
    model=Groq,
    response_format=ComplianceReview,
    system_prompt="""
You are Auriga's Brand Compliance Agent.

Review content for:

1. Professional tone
2. Misleading claims
3. Exaggeration
4. Technical accuracy
5. Brand consistency

If content is acceptable:
approved = true

Otherwise:
approved = false

Provide feedback.
"""
)
```

```python
response = compliance_agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": """
Title:
Why Agentic AI Is Transforming Enterprises

Content:
Agentic AI can help organizations automate
complex workflows and improve efficiency.

Auriga supports enterprises in exploring
AI adoption through consulting and digital
transformation initiatives.
"""
            }
        ]
    }
)
```

```python
review = response["structured_response"]

print(review.approved)
print(review.feedback)
print(review.risk_level)
```
