```python
from pydantic import BaseModel

class ContentStrategy(BaseModel):
    target_audience: str
    platform: str
    content_type: str
    angle: str
    call_to_action: str
```

```python
from dotenv import load_dotenv
load_dotenv()

from langchain_groq import ChatGroq

Groq = ChatGroq(
    model="llama-3.3-70b-versatile",
)
```

```python
from langchain.agents import create_agent

strategy_agent = create_agent(
    model=Groq,
    tools=[],
    response_format=ContentStrategy,
    system_prompt="""
    You are a Content Strategy Agent.

    Your responsibility is to:

    - Analyze trends
    - Analyze Auriga expertise
    - Select target audience
    - Select platform
    - Define content angle
    - Suggest CTA
    """
)
```

```python
response = strategy_agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": """
                Trend:
                Agentic AI

                Auriga Expertise:
                AI Consulting
                Data Analytics
                Digital Transformation
                """
            }
        ]
    }
)
```

```python
from pprint import pprint

pprint(response)
```
